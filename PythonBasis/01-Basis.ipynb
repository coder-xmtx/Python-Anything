# 1. variable 变量

四种基本数据类型

|strings 字符串 | Integer 整数 | Float 浮点数 | Boolean 布尔值|
|---|---|---|---|

```python
## strings 字符串
name = "Mixsu"

## Integer 整数
age = 20

## Float 浮点数
gpa = 4.5

## Boolean 布尔值
is_student = True

## F风格字符串打印
print(f"hello, I'm {name}, age {age}, GPA {gpa}. I'm a student, it's {is_student}.")
```

类型转换

|`str()` | `int()` | `float()` | `bool()` |
|---|---|---|---|

```python
## 浮点数转为整数
float_num = 2026.1
int_num = int(float_num)
print(int_num)
```
