# 7. format specifiers F风格字符串
`{value:flags}`

```python
price1 = 12.3
price2 = -13.14
price3 = 1520

# 保留x位小数 {price:.xf} / 千位逗号标识符 {price:,}
print(f"Price 1 : ${price1:.2f}")
print(f"Price 2 : ${price2:.2f}")
print(f"Price 3 : ${price3:,}")
print("------")

# 占有固定格数 {price:x} （x格）/ 占有固定格数且前面补零 {price:0x}
print(f"Price 1 : ${price1:10}")
print(f"Price 2 : ${price2:10}")
print(f"Price 3 : ${price3:010}")
print("------")

# 占有固定格数且左对齐 {price:<x}   同理：右对齐 {price:>x} 居中对齐 {price:^x}
print(f"Price 1 : ${price1:<10}")
print(f"Price 2 : ${price2:>10}")
print(f"Price 3 : ${price3:^10}")
print("------")

# 正数显示+号 {price:+} / 隐式+号 {price: }
print(f"Price 1 : ${price1:+}")
print(f"Price 2 : ${price2:+}")
print(f"Price 3 : ${price3: }")
print("------")

# 例子（组合）：price1 保留两位小数、右对齐、占有10格、前面添加+号
print(f"Special Price 1 : {price1:+10.2f}")
```

```python
# 例子：数字猜谜游戏（1~100）

low_number = 1
high_number = 100
random_number = random.randint(low_number, high_number)

is_running = True

count = 0

while is_running:
    guess_number = input("Enter your guess number: ")

    if not guess_number.isdigit():
        print("Please enter the number.")
    else:
        guess_number = int(guess_number)
        if guess_number < 1 or guess_number > 100:
            print("Your number is out of range.")
        elif guess_number < random_number:
            print(f"{guess_number} , Too Small.")
        elif guess_number > random_number:
            print(f"{guess_number} , Too Big")
        else:
            print(f"You're right! The right number is {random_number}.")
            is_running = False
    count += 1

print(f"You use {count} times.")
```

```python
# 例子：问候他人


def hello(first_name, last_name):
    print(f"Hello, {first_name} {last_name}.")


# 正确写法：
hello("Mingxiao", "Last")

# 调换过来就错了
hello("Last", "Mingxiao")
```

Default arguments 默认参数

可在函数头给某些参数定义默认值，这些参数要写在非默认参数后面

```python
# 例子：计算商品税后价


def net_price(list_price, discount=0, tax=0.05):
    """计算商品税后价，discount默认为0，tax默认为0.05"""
    return list_price * (1 - discount) * (1 + tax)


print(net_price(100))
```

Keyword arguments 关键字参数

可以显式地给对应的参数赋值，而不用按照位置顺序

注意：位置参数仍然排在最前面

```python
# 再次以上面的hello函数为例子

hello(first_name="Mingxiao", last_name="Last")
hello(last_name="Last", first_name="Mingxiao")
```

Arbitrary arguments 可变参数

在函数头写入`*args`代表接受可变位置参数，该参数在函数内表示为元组
- 不叫`*args`也行，反正要带个`*`

在函数头写入`**kwargs`代表接受可变关键字参数，该参数在函数内表示为字典
- 不叫`**kwargs`也行，反正要带个`**`

```python
# *args 例子

# 写一个函数，实现函数头写入多少数字都能相加


def add(a, b):  # 只能接受两个参数
    return a + b


def add_plus(*args):  # 可以接受多参数
    total = 0
    for arg in args:
        total += arg
    return total


print(add_plus(1, -1, 3, 5))
```

```python
# **kwargs 例子

# 写一个函数，可以打印家庭地址，无论写了多少有关家庭地址的前缀


def print_address(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}:{value}")


print_address(Country="China", Province="Guangdong", City="Guangzhou")
```

# 16. match case statement (like switch)

很像其他语言的 switch 语句，同样用 `case` 来表达情况，不过用 `_`（通配符）来表达else

```python
# 例子：数字来表示星期几


def day_of_week(day):
    match day:
        case 1:
            print("It is Monday.")
        case 2:
            print("It is Tuesday.")
        case 3:
            print("It is Wednesday.")
        case _:
            print("Not a valid day.")


day_of_week(1)
```

```python
# 例子：判断星期几是否为周末


def is_weekend(day):
    match day:
        case "Monday" | "Tuesday" | "Wednesday" | "Thursday" | "Friday":  # | 表示 or
            return False
        case "Saturday" | "Sunday":
            return True
        case _:
            return "Not a valid string"


print(is_weekend("haha"))
print(is_weekend("Saturday"))
```

multiple inheritance 多重继承

```python
# 例子：捕食者和猎物


class Animal_common:
    def __init__(self, name):
        self.name = name

    def eat(self):
        print(f"This {self.name} is eating.")

    def sleep(self):
        print(f"This {self.name} is sleeping.")


# 猎物
class Prey(Animal_common):
    def flee(self):
        print(f"This {self.name} is fleeing.")


# 捕食者
class Predator(Animal_common):
    def hunt(self):
        print(f"This {self.name} is hunting.")


class Rabbit(Prey):
    pass


class Hawk(Predator):
    pass


class Fish(Prey, Predator):
    pass


fish = Fish("Nemo")
fish.flee()  # 父类的方法
fish.eat()  # 父类的父类的方法
```

`super()` 拓展继承属性

如果子类和父类有一个同名方法，那么子类会覆盖父类

或者用`super().方法`使子类方法拓展父类方法

```python
# 例子：正方形、三角形、圆形都有颜色和填充属性，但是它们各有自己独特的属性


class Shape:
    def __init__(self, color, filled):
        self.color = color
        self.filled = filled

    def describe(self):
        print(f"It is {self.color} and {'filled' if self.filled else 'not fiiled'}  ")


class Circle(Shape):
    def __init__(self, color, filled, radius):
        super().__init__(color, filled)
        self.radius = radius

    def describe(self):
        super().describe()
        print(f"It is a circle with area of {3.14 * pow(self.radius, 2)}")


class Square(Shape):
    def __init__(self, color, filled, width):
        super().__init__(color, filled)
        self.width = width


class Triangle(Shape):
    def __init__(self, color, filled, width, height):
        super().__init__(color, filled)
        self.width = width
        self.height = height


circle = Circle("red", True, 5)
circle.describe()
```

duck typing 鸭子类型

鸭子类型在程序设计中是动态类型的一种风格。在这种风格中，一个对象有效的语义，不是由继承自特定的类或实现特定的接口，而是由"当前方法和属性的集合"决定。

可以这样理解：比如人和鸭子肯定是不一样的，鸭子会嘎嘎叫但是人不会，但是**叫声**这种东西，人和鸭子都可以发出，所以，假设鸭子类有一个`quack()`方法，人也可以定义一个`quack()`方法，这样子不使用继承也能够实现多态。

> If it looks like a duck and quacks like a duck, it must be a duck.

```python
# 鸭子类
class Duck:
    def quack(self):
        print("这鸭子正在嘎嘎叫")

    def feathers(self):
        print("这鸭子拥有白色和灰色的羽毛")


# 人类
class Person:
    def quack(self):
        print("这人正在模仿鸭子")

    def feathers(self):
        print("这人在地上拿起1根羽毛然后给其他人看")


# 函数/接口
def in_the_forest(duck):
    duck.quack()
    duck.feathers()


if __name__ == "__main__":
    donald = Duck()  # 创建一个Duck类的实例
    john = Person()  # 创建一个Person类的实例

    in_the_forest(donald)  # 调用函数，传入Duck的实例
    in_the_forest(john)  # 调用函数，传入Person的实例
```

static methods 静态方法

属于类而不是该类任何对象（实例）的方法，通常用于通用工具函数，其前需要`@staticmethod`装饰器

```python
class Employee:
    def __init__(self, name, position):
        self.name = name
        self.position = position

    def get_info(self):
        return f"{self.name} = {self.position}"

    @staticmethod
    def is_valid_position(position):
        valid_positions = ["Manager", "Cashier", "Cook", "Janitors"]
        return position in valid_positions


print(Employee.is_valid_position("Cook"))
```

class methods 类方法

`@classmethod` 声明类方法，参数`cls`指向的是类本身

- 静态方法和类方法的区别

    静态方法最适合不需要处理类级别数据使用，就当成一个类的工具
    
    类方法最适合在处理类级别数据时使用

```python
class Student:
    count = 0
    total_gpa = 0

    def __init__(self, name, gpa):
        self.name = name
        self.gpa = gpa
        Student.count += 1
        Student.total_gpa += gpa

    # INSTANCE METHOD
    def get_info(self):
        return f"{self.name} {self.gpa}"

    @classmethod
    def get_count(cls):
        return f"Total # of student: {cls.count}"

    @classmethod
    def get_average_gpa(cls):
        if cls.count == 0:
            return 0
        else:
            return f"{cls.total_gpa / cls.count:.2f}"


student01 = Student("Mixsu", 4.5)
student02 = Student("Vince", 3.9)


print(Student.get_count())
print(Student.get_average_gpa())
```

dunder methods 双下划线方法（类内置）

```python
class Book:
    def __init__(self, title, author, num_pages):
        self.title = title
        self.author = author
        self.num_pages = num_pages

    # 定义直接打印实例对象的方法（默认会返回该实例对象的内存地址）
    def __str__(self):
        return f"『 {self.title} 』 by {self.author}"

    # 定义两个实例相等的条件
    def __eq__(self, other):
        return self.title == other.title and self.author == other.author

    # 定义比较运算(自己定义的，不是内置)
    def __lt__(self, other):
        return self.num_pages < other.num_pages

    def __gt__(self, other):
        return self.num_pages > other.num_pages

    # 定义实例加法运算(自己定义的，不是内置)
    def __add__(self, other):
        return self.num_pages + other.num_pages

    # 定义包含运算
    def __contains__(self, keyword):
        return keyword in self.title


book1 = Book("Harry Potter and Philosopher's Stone", "J.K. Rowling", 223)
book2 = Book("The Lion, the Witch and the Wardrobe", "C.S. Lewis", 172)
book3 = Book("Harry Potter and Philosopher's Stone", "J.K. Rowling", 500)

print(book1)
print(book1 == book3)
print(book1 > book3)
print(book1 + book3)
print("Harry" in book1)
```

# 21. property decorator 属性装饰器

property 可以将一个方法定义为一个属性

```python
class Rectangle:
    def __init__(self, width, height):
        self._width = width
        self._height = height

    @property
    def width(self):
        return f"{self._width:.1f}"

    @property
    def height(self):
        return f"{self._height:.1f}"

    # 拓展属性赋值写法
    @width.setter
    def width(self, new_width):
        if new_width > 0:
            self._width = new_width
        else:
            print("Width must be than zero")

    @height.setter
    def height(self, new_height):
        if new_height > 0:
            self._height = new_height
        else:
            print("Width must be greater than zero")

    # 拓展删除属性写法
    @width.deleter
    def width(self):
        del self._width
        print("Width has been deleted.")

    @height.deleter
    def height(self):
        del self._height
        print("Height has been deleted.")


rectangle1 = Rectangle(3, 4)
rectangle1.width = 5

del rectangle1.width
```

decorators 装饰器

可以让其他函数挂载到基函数上，拓展原有函数功能

```python
def add_sprinkles(func):
    def wrapper(*args):
        print("You add the sprinkles.")
        func(*args)

    return wrapper


def add_fudge(func):
    def wrapper(*args):
        print("You add fudge.")
        func(*args)

    return wrapper


@add_sprinkles
@add_fudge
def get_ice_cream(flavor):
    print(f"Here is your {flavor} ice cream.")


get_ice_cream("villina")
```
