# 2. input 输入

`input()`函数可以让我们在终端输入文字或者值

注意：都会被转换为字符串类型

```python
# 例子：输入矩形的长和宽，输出面积
# 使用 float()进行类型转换

length = float(input("The length of area : "))
width = float(input("The width of area : "))
square = length * width

print(f"The square of area : {square}m²")
```

# 3. arithmetic and math 数学运算

加减乘除：`+` `-` `*` `/`

平方、取余：`**` `%`

num四舍五入x位小数、绝对值、a的b次方：`round(num,x)` `abs()` `power(a,b)`

在F风格字符串中，数字变量保留x位小数可以这样写：`print(f"We have {num:.xf} yuan.")`

--- `import math` ---

$\pi$、$e$、平方根：`math.pi` `math.e` `math.sqrt()`

向上取整、向下取整：`math.ceil()` `math.floor()`

```python
# 例子：计算三角形斜边长度，保留两位小数

import math

a, b = 3.5, 4.5
c = math.sqrt(pow(a, 2) + pow(b, 2))

print(f"The length of the hypotenuse of a triangle: {c:.2f}")
```

# 4. if else 条件判断

`if ... elif(else if)... else...`

注意 Python 的缩进规则

```python
# 例子：评价一个人的年纪

personAge = 20

if personAge <= 18:
    print("You are a kid.")
elif personAge <= 25:
    print("You are a youngth.")
elif personAge <= 50:
    print("You are an adult.")
else:
    print("You are an elder.")
```

conditional expressions 条件表达式（类似 三目运算符）

`X if condition else Y` => 如果满足 `condition` 返回 `X` ，否则返回 `Y`

```python
# 判断天气热不热

temperature_cold = 10

status = "COLD" if temperature_cold <= 10 else "HOT"

print(status)
```

# 5. logical operators 逻辑运算符

| `and` | `or` | `not`|
|---|---|---|
|全真则真|一真则真|非|

```python
# 例子：评价天气

temperature = 28
is_sunny = True

if temperature <= 10 and is_sunny:
    print(f"The temperature is {temperature}℃. The day is COLD and SUNNY.")
elif 10 < temperature < 26 and is_sunny:
    print(f"The temperature is {temperature}℃. The day is WARM and SUNNY.")
elif temperature >= 26 and is_sunny:
    print(f"The temperature is {temperature}℃. The day is HOT and SUNNY.")
else:
    print("Your input is not valid.")
```

# 6. string 字符串详解

string methods 字符串方法

```python
name = "mixsu code"

# 获取字符串长度 len()
print(len(name))  # 返回 10

# 查找第一个匹配 find()
print(name.find(" "))  # 匹配第一个空格的索引，返回 5
# 查找最后一个匹配 rfind()
print(name.rfind("e"))  # 匹配最后一个 e 的索引

# 第一个字母大写 capitalize()
print(name.capitalize())
# 全部大写 upper() / 全部小写 lower()
print(name.upper())

# 判断字符串是否为纯数字，是的话返回 True   isdigit()
print(name.isdigit())
# 判断字符串是否为纯字母（汉字也算），是的话返回 True   isalpha()
print(name.isalpha())

# 计数字符串某个字符 count()
print(name.count(" "))  # 空格的总数

# 替换字符串中某个字符 replace(old,new)
print(name.replace(" ", "-"))  # 将所有空格替换成-
```

```python
# 例子：用户名输入规则
# 1. 用户名不能超过12个字符
# 2. 用户名不能出现空格
# 3. 用户名不能出现数字

username = input("Enter your username : ")

if len(username) > 12:
    print("Your username can't be more than 12 characters!")
elif not username.find(" ") == -1:
    print("Your username can't contain spaces!")
elif not username.isalpha():
    print("Your username can't contain numbers!")
else:
    print(f"Welcome, {username}.")
```

string indexing 字符串索引 `[start : end : steps]` （区间左闭右开）

```python
telephone = "123-4567-8901"

print(telephone[0])  # 获取索引为 0 的字符
print(telephone[0:4])  # 获取索引为 0~3 的字符
print(telephone[5:])  # 获取索引为 5 到 末尾 的字符
print(telephone[-1])  # 获取倒数第 1 个字符
print(telephone[-4:])  # 获取倒数 4 个字符
print(telephone[::2])  # 获取全部字符的情况下，步长为 2
print(telephone[::-1])  # 反转字符串
```

# 7. format specifiers F风格字符串
`{value:flags}`

```python
price1 = 12.3
price2 = -13.14
price3 = 1520

# 保留x位小数 {price:.xf} / 千位逗号标识符 {price:,}
print(f"Price 1 : ${price1:.2f}")
print(f"Price 2 : ${price2:.2f}")
print(f"Price 3 : ${price3:,}")
print("------")

# 占有固定格数 {price:x} （x格）/ 占有固定格数且前面补零 {price:0x}
print(f"Price 1 : ${price1:10}")
print(f"Price 2 : ${price2:10}")
print(f"Price 3 : ${price3:010}")
print("------")

# 占有固定格数且左对齐 {price:<x}   同理：右对齐 {price:>x} 居中对齐 {price:^x}
print(f"Price 1 : ${price1:<10}")
print(f"Price 2 : ${price2:>10}")
print(f"Price 3 : ${price3:^10}")
print("------")

# 正数显示+号 {price:+} / 隐式+号 {price: }
print(f"Price 1 : ${price1:+}")
print(f"Price 2 : ${price2:+}")
print(f"Price 3 : ${price3: }")
print("------")

# 例子（组合）：price1 保留两位小数、右对齐、占有10格、前面添加+号
print(f"Special Price 1 : {price1:+10.2f}")
```

# 8. while loops （while循环）

```python
# 例子：输入1~10之间的数字，如果输入不正确则重新输入
input_number = int(input("Enter a number between 1 ~ 10 : "))

while input_number < 1 or input_number > 10:
    print(f"{input_number} is not valid.")
    input_number = int(input("Enter a number between 1 ~ 10 : "))

print(f"Your number is {input_number}.")
```
