# 6. string 字符串详解

string methods 字符串方法

```python
name = "mixsu code"

# 获取字符串长度 len()
print(len(name))  # 返回 10

# 查找第一个匹配 find()
print(name.find(" "))  # 匹配第一个空格的索引，返回 5
# 查找最后一个匹配 rfind()
print(name.rfind("e"))  # 匹配最后一个 e 的索引

# 第一个字母大写 capitalize()
print(name.capitalize())
# 全部大写 upper() / 全部小写 lower()
print(name.upper())

# 判断字符串是否为纯数字，是的话返回 True   isdigit()
print(name.isdigit())
# 判断字符串是否为纯字母（汉字也算），是的话返回 True   isalpha()
print(name.isalpha())

# 计数字符串某个字符 count()
print(name.count(" "))  # 空格的总数

# 替换字符串中某个字符 replace(old,new)
print(name.replace(" ", "-"))  # 将所有空格替换成-
```

```python
# 例子：用户名输入规则
# 1. 用户名不能超过12个字符
# 2. 用户名不能出现空格
# 3. 用户名不能出现数字

username = input("Enter your username : ")

if len(username) > 12:
    print("Your username can't be more than 12 characters!")
elif not username.find(" ") == -1:
    print("Your username can't contain spaces!")
elif not username.isalpha():
    print("Your username can't contain numbers!")
else:
    print(f"Welcome, {username}.")
```

string indexing 字符串索引 `[start : end : steps]` （区间左闭右开）

```python
telephone = "123-4567-8901"

print(telephone[0])  # 获取索引为 0 的字符
print(telephone[0:4])  # 获取索引为 0~3 的字符
print(telephone[5:])  # 获取索引为 5 到 末尾 的字符
print(telephone[-1])  # 获取倒数第 1 个字符
print(telephone[-4:])  # 获取倒数 4 个字符
print(telephone[::2])  # 获取全部字符的情况下，步长为 2
print(telephone[::-1])  # 反转字符串
```

# 7. format specifiers F风格字符串
`{value:flags}`

```python
price1 = 12.3
price2 = -13.14
price3 = 1520

# 保留x位小数 {price:.xf} / 千位逗号标识符 {price:,}
print(f"Price 1 : ${price1:.2f}")
print(f"Price 2 : ${price2:.2f}")
print(f"Price 3 : ${price3:,}")
print("------")

# 占有固定格数 {price:x} （x格）/ 占有固定格数且前面补零 {price:0x}
print(f"Price 1 : ${price1:10}")
print(f"Price 2 : ${price2:10}")
print(f"Price 3 : ${price3:010}")
print("------")

# 占有固定格数且左对齐 {price:<x}   同理：右对齐 {price:>x} 居中对齐 {price:^x}
print(f"Price 1 : ${price1:<10}")
print(f"Price 2 : ${price2:>10}")
print(f"Price 3 : ${price3:^10}")
print("------")

# 正数显示+号 {price:+} / 隐式+号 {price: }
print(f"Price 1 : ${price1:+}")
print(f"Price 2 : ${price2:+}")
print(f"Price 3 : ${price3: }")
print("------")

# 例子（组合）：price1 保留两位小数、右对齐、占有10格、前面添加+号
print(f"Special Price 1 : {price1:+10.2f}")
```

```python
# 例子：制作一个简易的倒计时（时分秒）
# 用户输入秒数，输出倒计时（00：00：00）

import time

input_time = int(input("Enter the time in seconds : "))

for t in range(input_time, 0, -1):
    seconds = t % 60
    minutes = int(t / 60) % 60
    hours = int(t / 3600)
    print(f"\r{hours:02}:{minutes:02}:{seconds:02}")
    time.sleep(1)  # 延迟1秒执行

print("Done.")
```

# 11. dictionary 字典（键值对）

```python
# 例子：数字猜谜游戏（1~100）

low_number = 1
high_number = 100
random_number = random.randint(low_number, high_number)

is_running = True

count = 0

while is_running:
    guess_number = input("Enter your guess number: ")

    if not guess_number.isdigit():
        print("Please enter the number.")
    else:
        guess_number = int(guess_number)
        if guess_number < 1 or guess_number > 100:
            print("Your number is out of range.")
        elif guess_number < random_number:
            print(f"{guess_number} , Too Small.")
        elif guess_number > random_number:
            print(f"{guess_number} , Too Big")
        else:
            print(f"You're right! The right number is {random_number}.")
            is_running = False
    count += 1

print(f"You use {count} times.")
```

# 13. function 函数

**写法：**`def functionName():`

Position arguments 位置参数

默认写法，每一个位置对应一个值，不可随意调换

```python
# 例子：问候他人


def hello(first_name, last_name):
    print(f"Hello, {first_name} {last_name}.")


# 正确写法：
hello("Mingxiao", "Last")

# 调换过来就错了
hello("Last", "Mingxiao")
```

Default arguments 默认参数

可在函数头给某些参数定义默认值，这些参数要写在非默认参数后面

```python
# 例子：计算商品税后价


def net_price(list_price, discount=0, tax=0.05):
    """计算商品税后价，discount默认为0，tax默认为0.05"""
    return list_price * (1 - discount) * (1 + tax)


print(net_price(100))
```

Keyword arguments 关键字参数

可以显式地给对应的参数赋值，而不用按照位置顺序

注意：位置参数仍然排在最前面

```python
# 再次以上面的hello函数为例子

hello(first_name="Mingxiao", last_name="Last")
hello(last_name="Last", first_name="Mingxiao")
```

Arbitrary arguments 可变参数

在函数头写入`*args`代表接受可变位置参数，该参数在函数内表示为元组
- 不叫`*args`也行，反正要带个`*`

在函数头写入`**kwargs`代表接受可变关键字参数，该参数在函数内表示为字典
- 不叫`**kwargs`也行，反正要带个`**`

```python
# *args 例子

# 写一个函数，实现函数头写入多少数字都能相加


def add(a, b):  # 只能接受两个参数
    return a + b


def add_plus(*args):  # 可以接受多参数
    total = 0
    for arg in args:
        total += arg
    return total


print(add_plus(1, -1, 3, 5))
```

```python
# **kwargs 例子

# 写一个函数，可以打印家庭地址，无论写了多少有关家庭地址的前缀


def print_address(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}:{value}")


print_address(Country="China", Province="Guangdong", City="Guangzhou")
```

# 15. list comprehensions 列表推导式

for循环的另外一种写法，能够快速创建列表，遵循以下写法

`[expression for value in iterable if condition]`

```python
# 例子：生成一个1~10数字列表，且将其中每个数字翻倍

# for循环写法
double_list1 = []

for num in range(1, 11):
    double_list1.append(num * 2)

print("FOR循环", double_list1)

# 列表推导式写法
double_list2 = [x * 2 for x in range(1, 11)]

print("列表推导式", double_list2)
```

```python
# 例子：筛选出给定数字列表中的正数（列表推导式）

number_list = [1, -2, 3, -4, 5, -6]

positive_number_list = [num for num in number_list if num > 0]

print(positive_number_list)
```

# 16. match case statement (like switch)

很像其他语言的 switch 语句，同样用 `case` 来表达情况，不过用 `_`（通配符）来表达else

```python
# 例子：数字来表示星期几


def day_of_week(day):
    match day:
        case 1:
            print("It is Monday.")
        case 2:
            print("It is Tuesday.")
        case 3:
            print("It is Wednesday.")
        case _:
            print("Not a valid day.")


day_of_week(1)
```

```python
# 例子：判断星期几是否为周末


def is_weekend(day):
    match day:
        case "Monday" | "Tuesday" | "Wednesday" | "Thursday" | "Friday":  # | 表示 or
            return False
        case "Saturday" | "Sunday":
            return True
        case _:
            return "Not a valid string"


print(is_weekend("haha"))
print(is_weekend("Saturday"))
```
