# 2. 索引与切片

```python
import numpy as np
```

## 2.1 一维数组的索引与切片

```python
arr = np.random.randint(1,100,20)
```

```python
print(arr[10]) # 索引
print(arr[:]) # 全部
print(arr[2:5]) # 切片（左闭右开）
print(arr[slice(2,15,3)]) # 切片函数 slice(start,end,step)
print(arr[(arr>50) & (arr<70)]) # 过滤
```

## 2.2 二维数组的索引和切片

```python
arr2 = np.random.randint(1,100,(4,8))
arr2
```

```python
print(arr2[1,3]) # 索引1【注意起始为(0,0)】
print(arr2[1][3]) # 索引2【注意起始为(0,0)】
print(arr2[:,:]) # 全部
print(arr2[2]) # 获取行
print(arr2[:,3]) # 获取列
print(arr2[1,2:5]) # 第1行，第2个到第4个
print(arr2[arr2>50]) # 过滤，输出一维数组
print(arr2[2][arr2[2]>50]) # 过滤第2行，挑出大于50的
```
