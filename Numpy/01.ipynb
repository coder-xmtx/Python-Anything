# 1. **ndarray**

```python
import numpy as np
```

## 1.1 属性

|属性|解释|
|---|---|
|`shape`|数组的形状（行数和列数）|
|`ndim`|维度数量：数组是几维的|
|`size`|总元素个数|
|`dtype`|元素类型|

```python
list1 = [[1,2,3],[4,5,6]]
arr = np.array(list1,dtype=np.float64)
arr
```

## 1.2 方法
1. `arr.copy()`可以拷贝数组，默认为深拷贝，即拷贝后的新数组和原数组没什么联系，修改新数组不会改变原数组

```python
arr_copy = arr.copy()
arr_copy
```

## 1.3 生成特殊一维数组

2. `np.arange(start,end,step)`生成等差数列，左闭右开

```python
arithmetic_sequence = np.arange(1,10,1)
arithmetic_sequence
```

3. `np.linspace(start,end,per)`生成等间隔数列，第三个参数`per`写入要等分的份数，左闭右闭

```python
linspace = np.linspace(1,10,5)
linspace
```

4. `np.logspace(start,end,per,base=底数)`生成等比数列，最后一个参数写底数

```python
logspace = np.logspace(0,4,3,base=2,dtype=int)
logspace
```

## 1.4 特殊矩阵

1. `np.eye(num)`定义单位矩阵，`num`值为行列

```python
identity_matrix = np.eye(3,dtype=int)
identity_matrix
```

2. `np.diag([num1,num2,...])`定义对角矩阵，`num`值为主对角线的值

```python
diagonal_matrix = np.diag([1,2,3])
diagonal_matrix
```

3. `np.full((x,y),num)`可以生成指定行列的二维数组，并将数组全部填充`num`

```python
arr_grow = np.full((3,3),0)
arr_grow
```

## 1.5 随机数组

1. `random.rand(shape)`生成0~1之间的随机浮点数（均匀分布）

```python
random1 = np.random.rand(2,3)
random1
```

2. `random.uniform(start,end,(shape))`生成指定范围的随机浮点数（均匀分布）

```python
random2 = np.random.uniform(3,6,(2,3))
random2
```

3. `random.randint(start,end,(shape))`生成指定范围的随机整数（均匀分布）

```python
random3 = np.random.randint(2,10,(3,4))
random3
```

4. `random.randn(shape)`，生成随机浮点数数组（标准正态分布）

```python
random4 = np.random.randn(2,3)
random4
```

## 1.6 设置随机种子

#### 为什么需要随机种子
在生成随机数组时，每次生成的结果都不一样，但如果我们需要将结果复现给其他人看，不一样的随机数据最后呈现出来的肯定不一样，
所以为了保证复现一致，可以设置随机种子来保证生成的随机数组具有唯一性。

#### 设置方法
直接写 `np.random.seed(num)`，填入数字即可，数字可以随便填

```python
np.random.seed(42) # 直接写，不需要赋给变量

random_array = np.random.randint(2,10,(3,4))
random_array
```
