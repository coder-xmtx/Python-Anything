# Q1: 学生成绩统计
创建一个包含10名学生数学成绩的Series，成绩范围在50-100之间。<br />
计算平均分、最高分、最低分，并找出高于平均分的学生人数。

```python
# 创建 Series
np.random.seed(1)
Series01 = np.random.randint(50, 101, 10)
Series01 = pd.Series(Series01,index=['student No.'+ str(i) for i in range(1,11)])

# 平均分、最高分、最低分
print(Series01)
print(f"平均分：{Series01.mean()}  最高分：{Series01.max()}  最低分：{Series01.min()}")

# 高于平均分的学生人数
print(f"高于平均分的人数：{Series01[Series01 > Series01.mean()].count()}")
```

# Q2：城市温度分析
给定某城市一周每天的最高温度Series，完成以下任务：
- 找出温度超过30度的天数
- 计算平均温度
- 将温度从高到低排序
- 找出温度变化最大的相邻天

```python
# 温度数据
temperatures = pd.Series([28, 31, 29, 32, 30, 27, 33],index=['周一','周二','周三','周四','周五','周六','周日'])
# 前三个小问分析
print(f"温度超过30度的天数：{temperatures[temperatures>=30].size}")
print(f"平均温度：{temperatures.mean():.2f}")
print(f"高到低排序：\n{temperatures.sort_values(ascending=False)}")
# 温度变化最大的相邻天
max_change_day = temperatures.diff().abs().idxmax()
prev_day = temperatures.index[temperatures.index.get_loc(max_change_day)-1]
print(f"温度变化最大的相邻天：{prev_day} 到 {max_change_day}")
```

# Q3: 股票价格分析
给定某股票连续10个交易日的收盘价Series：
- 计算每日收益率（当日收盘价/前日收盘价-1）
- 找出收益率最高和最低的日期
- 计算波动率（收益率的标准差）

```python
# 数据
prices = pd.Series([102.3, 103.5, 105.1, 104.8, 106.2, 107.0, 106.5, 108.1, 109.3, 110.2],
                   index=pd.date_range('2025-12-01', periods=10,freq='D'))
# 分析
benifit = prices.pct_change()
# 收益率最高和最低的日期
benifit_max = benifit.idxmax()
benifit_min = benifit.idxmin()
# 波动率
benifit_std = benifit.std()

print(f"每日收益率：\n{benifit}")
print(f"收益率最高那天：{benifit_max} 最低那天：{benifit_min} 波动率：{benifit_std:.4f}")
```

# Q4: 销售数据分析
某产品过去12个月的销售量Series：
- 计算季度平均销量（每3个月为一个季度）
- 找出销量最高的月份
- 计算月环比增长率
- 找出连续增长超过2个月的月份

```python
# 数据
sales = pd.Series([120, 135, 145, 160, 155, 170, 180, 175, 190, 200, 210,220],index=pd.date_range('2022-01-01', periods=12, freq='ME'))
# 季度平均销量
sales_season = sales.resample('QS').mean()
# sales_season.index = [f"Q{i}" for i in range(1, len(sales_season)+1)]
sales_season.index = ['Q1','Q2','Q3','Q4']
# 销量最高的月份
sales_max = sales.idxmax()
# 月环比增长率
sales_benefit = sales.pct_change()

# 找出连续增长超过2个月的月份
## 可以理解为月增长率不为0的月份大于等于3，所以可以先用布尔索引找出正增长的月份，
## 再用窗口函数不断扫过连续3个月，找出三个True，即为3
sales_up = sales_benefit > 0
sales_up_3 = sales_up[sales_up.rolling(3).sum()==3].keys()

# 结果
print(f"季度平均销量：\n{sales_season}\n")
print(f"销售最高的月份：\n{sales_max}\n")
print(f"月环比增长率：\n{sales_benefit}\n")
print(f"连续增长超过2个月的月份：\n{[d.strftime('%m') for d in sales_up_3]}")
```

# Q5: 每小时销售数据分析
某商店每小时销售额Series:
- 按天重采样计算每日总销售额
- 计算每天营业时间（8:00-22:00）和非营业时间的销售额比例
- 找出销售额最高的3个小时

```python
# 数据
np.random.seed(42) # 设置随机种子
hourly_sales = pd.Series(np.random.randint(0, 100, 24),
                         index=pd.date_range('2025-01-01', periods=24, freq='h'))

# 重采样计算总销售额
day_sum =  hourly_sales.resample('D').sum().sum()
# 另一种写法：day_sum =  hourly_sales.sum()
# 为什么会要加两个sum，可以发现，只加一个sum，重采样后得到的还是一个Series，只不过存了总和的数据，再加一个sum就是纯数字

# 营业时间和非营业时间销售比例
## 得到所需数据的方法：
### 方法一：between_time 函数
work_sales = hourly_sales.between_time('8:00','22:00')
### 方法二：布尔索引
hourly_sales[(hourly_sales.index.hour >= 8) & (hourly_sales.index.hour <= 22)]
### 方法三：切片
hourly_sales.iloc[8:23]
## 计算
work_sales_percent = work_sales.sum() / day_sum

# 销售额最高的三个小时
sales_maxhour_3 = hourly_sales.nlargest(3)

# 总结果输出
print(f"总销售额：\n{day_sum}\n")
print(f"营业时间与非营业时间销售比例：\n营业 {work_sales_percent * 100 :.3f}% \n非营业 {(1-work_sales_percent) * 100 :.3f}%\n")
print(f"销售额最高的三个小时：\n{sales_maxhour_3.index.hour}")
```
